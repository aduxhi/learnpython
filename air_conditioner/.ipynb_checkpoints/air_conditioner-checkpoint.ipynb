```python
import pandas as pd;
import matplotlib.pyplot as plt;
from pandas import Series, DataFrame;
import matplotlib.pyplot as plt;
```

```python
oridata = pd.read_csv('air_conditioner.csv', index_col = None, encoding='gbk',)
oridata.head()
```

```python
plt.figure(figsize=(10,10))
plt.plot(oridata['ia'][1:1440*7],)
plt.plot('ia')
```

```python
from scipy import signal
import numpy as np
import copy
```

```python
ia = oridata['ia']
iaa = copy.deepcopy(ia)

for i in range(5):
    peaks, _ = signal.find_peaks(iaa, height = ia.mean(), width=[0,4], prominence=8)
    iaa[peaks] = iaa[peaks-1]
plt.figure(figsize=(10,10))
plt.subplot(211)
plt.plot(ia)
plt.subplot(212)
plt.plot(iaa)
```

```python
s = iaa[:1440]
x = np.quantile(s, 0.95)
plt.figure(figsize=(10,10))
plt.subplot(211)
plt.plot(s)
plt.axhline(x)
```
